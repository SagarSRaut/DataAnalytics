```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import os
```

```python
df = pd.read_csv('tips.csv')
df
```

```python
#plotting distributions on total_bill column

total_bill_series = df['total_bill']
# Calculate mean and standard deviation
mu, std = total_bill_series.mean(), total_bill_series.std()

# Calculate z-score for each value
z_scores = (total_bill_series - mu) / std

# Plot normal distribution graph of BloodPressure
plt.figure(figsize=(12, 6))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(total_bill_series, bins=20, density=True, alpha=0.6, color='b', edgecolor='black')
```
