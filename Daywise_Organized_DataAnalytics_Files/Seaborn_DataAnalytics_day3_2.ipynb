```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv("dm_office_sales.csv")
print(df.head())
print(df.info())
sns.scatterplot(x='salary',y='sales',data=df)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(16, 12))

plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df ,hue='division',ax=axs[0, 0])


plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df ,hue='work experience',ax=axs[0, 1])


plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df ,hue='work experience',palette='viridis',ax=axs[1, 0])


axs[1, 1].axis('off')

plt.tight_layout()

plt.show()
# axes[0][1].plot(x,y)
# axes[1][0].plot(y,x)
```

```python
#KDE - (kernel density estimation) plot
```

```python
import numpy as np
np.random.seed(42)
sample_ages= np.random.randint(0,100,200)
print(sample_ages)

sample_ages = pd.DataFrame(sample_ages,columns=['age'] )
print(sample_ages.head())

sns.rugplot(data=sample_ages, x='age')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages, x='age',bins=10,rug=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.displot(data=sample_ages, x='age',bins=10,rug=True,kde=True)
plt.show()
```

```python
sns.kdeplot(data=sample_ages, x='age')
plt.show() #Density given in the graph is the probablity
```
