```python
#conditions
df = df.reset_index()
print(df['total_bill']>30) #returns boolean values 
```

```python
bool_series = df['total_bill'] > 30
print(bool_series) #returns boolean
print(df[bool_series]) #returns the actual values  at all 'true' places
```

```python
print(df[df['total_bill']>30]) #same output as above but this is a direct syntax
```

```python
print(df.total_bill > 30) # same as df['total_bill']>30]
```

```python
print(df[df.total_bill > 30]) # same as df[df['total_bill']>30]
```

```python
df[df['gender'] == 'Female']
```

```python
#Multiple conditions
df_new = df[(df['total_bill'] > 30) & (df['gender'] == 'Male')]
df_new
```

```python
df_new = df[(df['total_bill'] > 30) & ~(df['gender'] == 'Female')]
df_new
```

```python
df_new = df[(df['total_bill'] > 30) & (df['gender'] != 'Female')]
df_new
```

```python
df_new = df[(df['total_bill'] > 30) | (df['tip'] > 5)]
df_new
```

```python
# conditinal is in operator
options = ['Sat','Sun']
df['day'].isin(options) # this is same as df[df['day'].isin(['Sat','Sun'])])
```
