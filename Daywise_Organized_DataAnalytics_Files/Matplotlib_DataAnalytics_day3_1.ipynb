```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic array plot
x = np.arange(0,10)
y = 2*x
print(x)
print(y)
```

```python
#Line plot
plt.xlabel('X-axis title')
plt.ylabel('Y-axis title')
plt.title('Plot title here')
plt.plot(x,y)
#plt.show #required for non-jupyter users, but also removes out[] info
```

```python
# More features
plt.plot(x,y)
plt.xlabel('X-axis title')
plt.ylabel('Y-axis title')
plt.title('Plot title here')
plt.xlim(0,6) #lower limit, upper limit
plt.ylim(0,12) #lower limit, upper limit
plt.show()
```

```python
#Exporting a plot
plt.plot(x,y)
plt.savefig('example.png')
```

```python
a=np.linspace(0,10,11)
b=a**4
x=np. arange(0,10)
y=2*x

fig=plt.figure()
axes1= fig.add_axes([0,0,1,1])   # LArge figure
axes1.plot(x,y)
axes1.set_xlabel('X label') 
axes1.set_ylabel('Y label') 
axes1.set_title('Big figure')


axes2=fig.add_axes([0.2,0.5,0.25,0.25])  # smaller figure
axes2.plot(a,b)
axes2.set_xlim(8,10) 
axes2.set_ylim(4000,10000) 
axes2.set_xlabel('X') 
axes2.set_ylabel('Y') 
axes2.set_title('Zoomed In')


plt.show()
```

```python
a=np.linspace(0,10,11)
b=a**4

x=np. arange(0,10)
y=2*x

fig,axes =plt.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');  #; hides out[]
plt.show()

```

```python
# Empty canvas of 1 by 2 subplots

fig,axes =plt.subplots(nrows=1,ncols=2)

print(axes)
print(axes.shape)
plt.show()
```

```python
#empty canvas of 2 by 2 subplot

fig,axes =plt.subplots(nrows=2,ncols=2)
print(fig)
print(axes)
print(axes.shape)
plt.show()
```

```python
# used .subplot adjust to adjust spacing manually

fig,axes =plt.subplots(nrows=2,ncols=2 ,figsize=(12,8))
axes[0][0].plot(a,b)
axes[0][1].plot(x,y)
axes[1][1].plot(y,x)
axes[1][0].plot(b,a)
fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.9,hspace=0.1,)
plt.show()
 
```

```python
#use left, right, top, bottom to stretch subplots
#use wspace, hspace to add spacing btween subplots

fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.9,hspace=0.1,)
plt.show()
```

```python
#Legends

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('x**2  vs  x**3')


ax.plot(x,x**2,label='x**2') #label is the attribute used for the legend
ax.plot(x,x**3,label='x**3')
# ax.legend() #the legend method is called
# ax.legend(loc=1) --> legend is shifted towards top-right corner
# ax.legend(loc=2) --> legend is shifted towards top-left corner
# ax.legend(loc=3) --> legend is shifted towards bottom-left corner
# ax.legend(loc=4) --> legend is shifted towards bottom-right corner

plt.show()
```

```python
#Setting colours, line widths, linetypes
#MATLAB style line color and style

fig,ax = plt.subplots()
ax.plot(x,x**2,'b*-') #blue line with dots
ax.plot(x,x**3,'g--') #green dashed line
plt.show()
```

```python
#Suggested approach: use keyword arguments

fig,ax = plt.subplots()
ax.plot(x,x+1,color='blue',alpha=0.5) #blue line with dots
ax.plot(x,x+2,color='#3F51B5') #hex code
ax.plot(x,x+3,color='#E91E63') #hex code
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,6))
ax.plot(x,x-1,color='orange',lw=10) #blue line with dots
ax.plot(x,x-2,color='white',lw=10) #hex code
ax.plot(x,x-3,color='green',lw=10) #hex code
plt.show()
```

```python
# Linestyles
fig,ax = plt.subplots(figsize=(12,6))
ax.plot(x,x-1,color='orange',lw=10,linestyle='-') #solid
ax.plot(x,x-2,color='red',lw=10,ls='-.') #dash and dot
ax.plot(x,x-3,color='green',lw=10,ls=':') #dots
ax.plot(x,x-4,color='black',lw=10,ls='--') #dashes
plt.show()
```

```python
# Markers
fig,ax = plt.subplots(figsize=(12,6))
ax.plot(x,x-1,marker='+',markersize=20) #solid
ax.plot(x,x-2,marker='o',ms=20) #dash and dot
ax.plot(x,x-3,marker='s', ms =20,lw=0) 
ax.plot(x,x-4,marker='1',ms=20) #dashes
plt.show()
```
